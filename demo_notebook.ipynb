```python
folder_to_scan = "./solvers"

import hashlib
import os

def get_file_hash(file_path):
    """Calculate SHA-256 hash of a single file."""
    sha256_hash = hashlib.sha256()
    try:
        with open(file_path, "rb") as f:
            # Read file in 64kb chunks to handle large files
            for byte_block in iter(lambda: f.read(65536), b""):
                sha256_hash.update(byte_block)
        return sha256_hash.hexdigest()
    except Exception as e:
        return f"Error: {e}"

def hash_folder_files(directory_path):
    """Iterate through a folder and print the hash of each file."""
    if not os.path.isdir(directory_path):
        print(f"Error: {directory_path} is not a valid directory.")
        return

    print(f"{'File Name':<30} | {'SHA-256 Hash'}")
    print("-" * 100)

    # Use os.listdir for files in the immediate folder
    # Use os.walk(directory_path) if you need to include subfolders
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        
        # Only process files, skip directories
        if os.path.isfile(file_path):
            file_hash = get_file_hash(file_path)
            print(f"{filename:<30} | {file_hash}")

# Replace with your actual folder path
hash_folder_files(folder_to_scan)
```
